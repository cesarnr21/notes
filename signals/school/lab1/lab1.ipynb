```python
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
import os
from scipy.io import wavfile

data_path = Path(os.getcwd()).parent.parent / 'data'

# load voice recordings
# voice 1 female, word 2 male
# NOTE: at least here, all of these files has the same sample rate of 44.1kHz
samp_rate, female_1 = wavfile.read(data_path / 'university1.wav')
samp_rate, male_1 = wavfile.read(data_path / 'university2.wav')

samp_rate, female_2 = wavfile.read(data_path / 'villanova1.wav')
samp_rate, male_2 = wavfile.read(data_path / 'villanova2.wav')

# samp_rate, self_university = wavfile.read(data_path / 'university_self.wav')
# samp_rate, self_villanova = wavfile.read(data_path / 'villanova_self.wav')

# NOTE: the signal values are a 2-d array since they are meant for a stereo speaker
# simply cut one column from each array
female_1 = female_1[:,0]
female_2 = female_2[:,0]
# male_1 = male_1[:,0]
# male_2 = male_2[:,0]
# self_university = self_university[:,0]
# self_villanova = self_villanova[:,0]
```

```python
# will also reuse samp_period and time for plot
samp_period = 1 / samp_rate 
time = np.linspace(0, len(female_1) * samp_period, len(female_1))

# Plot Female Recordings
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(10, 10))
fig.tight_layout(pad=5)
fig.suptitle('female voice')

ax1.plot(time, female_1)
ax1.set(title='"university" recording time plot', xlabel='time [seconds]', ylabel='amplitude')

ax3.plot(time, female_2)
ax3.set(title='"villanova" recording time plot', xlabel='time [seconds]', ylabel='amplitude')

ax2.specgram(female_1, mode= 'psd', scale='dB')
ax2.set(title='"university" spectrogram', xlabel='frequency [Hz]', ylabel='power density')

ax4.specgram(female_2)
ax4.set(title='"villanova" spectrogram', xlabel='frequency [Hz]', ylabel='power density')
```
